# EDA Analysis

This notebook displays the generated EDA chart outputs from `reports/charts/` and provides key findings for the final report.

```python
from pathlib import Path
from IPython.display import display, Image, HTML, IFrame

root = Path('.')
chart_dir = root / "reports" / "charts"
png_files = sorted(chart_dir.glob("*.png"))
html_files = sorted(chart_dir.glob("*.html"))

if not chart_dir.exists():
    raise FileNotFoundError(f"Chart directory not found: {chart_dir}")

print(f"Found {len(png_files)} PNG charts and {len(html_files)} HTML fallback charts in {chart_dir}")

for path in png_files:
    display(HTML(f"<h3>{path.name}</h3>"))
    display(Image(filename=str(path), width=900))

for path in html_files:
    display(HTML(f"<h3>{path.name}</h3>"))
    display(IFrame(src=str(path), width=1100, height=650))
```

## Key Findings

1. NAV trends show strong growth for top funds from 2022 through 2025 with a visible late 2024 correction phase.
2. AUM growth remains concentrated in the largest fund houses, with the top 10 dominating overall market share.
3. Monthly SIP inflows trend upward, indicating solid investor commitment and systematic investment behavior.
4. Heatmap analysis reveals distinct category-level inflow patterns, with equity categories leading overall transactions.
5. Investor demographics highlight an age-skew toward younger and middle-aged segments, with a balanced gender distribution.
6. Tier 1 cities contribute disproportionally to fund flows, showing geographic concentration in urban centers.
7. Total NAV return correlations among top funds suggest strong co-movement, especially within the same categories.
8. Sector allocation is heavily weighted in financials and technology, per the top holdings sector donut.
9. Active SIP account growth outpaces new SIP account creation, signaling increasing portfolio consolidation amongst existing investors.
10. SIP AUM trends reinforce resilience in systematic investment volumes even during moderate NAV volatility.
